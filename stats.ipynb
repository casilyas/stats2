```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
import seaborn as sns
df =pd.read_csv("w5_Telecom_customer churn.csv")
df.dropna(inplace=True)
for c in df.columns[:4]:
    print(c)
    q3=df[c].quantile(0.75)
    q1=df[c].quantile(0.25)
    iqr=q3-q1
    print("mean:",df[c].mean(),
          "\n","mode:",df[c].mode().iloc[0],
          "\n","median:",df[c].median(),
          "\n","q3:",q3,"q1:",q1,"iqr:",iqr,
         "\n","min:",df[c].min(),"max:",df[c].max(),
          "\n","skew:",df[c].skew(),
          "\n","kutosis:",kurtosis(df[c])
         )
    sns.boxplot(df[c])
    plt.show()
#jb_stat, p_value = stats.jarque_bera(df[:4])
#print("Jarque-Bera statistic:", jb_stat)
#print("p-value:", p_value) 
print(df[:4].corr())
```
